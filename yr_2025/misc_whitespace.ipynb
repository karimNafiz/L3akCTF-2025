```python
# this block will be used to exclude the combinations that have the special characters

def filter_possible_combos(special_char_set , list_possible_combos):
    new_table = []
    for x in list_possible_combos:
        new_list = []
        for y in x:
            flag = False 
            for z in y:
                if z in special_char_set:
                    flag = True 
                    break
            if not flag:
                new_list.append(y)
        new_table.append(new_list)
    return new_table

list_possible_combos_filtered = filter_possible_combos(special_char_set , list_possible_combos)
for x in list_possible_combos_filtered:
    print(x)

```
